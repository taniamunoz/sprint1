```python
class Producto:
    #Constructor de la clase producto
    def __init__(self, id, nombre, descripcion, cantidad, precio):
        self.id = id
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

    #funcion para imprimir los productos
    def __str__(self):
        return f"ID: {self.id}, Nombre: {self.nombre}, Descripción: {self.descripcion}, Cantidad: {self.cantidad}, Precio: {self.precio}"
    
    #metodo para obtener el id
    def obtener_id_producto(self):
        return self.id

    #metodo para obtener el nombre
    def obtener_nombre_producto(self):
        return self.nombre

    #metodo para actualizar el nombre
    def set_nombre_producto(self, value):
        self.nombre = value

    #metodo para obtener la descripcion
    def obtener_descripcion_producto(self):
        return self.descripcion

    #metodo para actualizar la descripcion
    def set_descripcion_producto(self, value):
        self.descripcion = value
        
    #metodo para obtener la cantidad
    def obtener_cantidad_producto(self):
        return self.cantidad

    #metodo para actualizar la cantidad
    def set_cantidad_producto(self, value):
        self.cantidad = value

    #metodo para obtener el precio
    def obtener_precio_producto(self):
        return self.precio

    #metodo para actualizar el precio
    def set_precio_producto(self, value):
        self.precio= value
```

```python
class Inventario:
    #Constructor de la clase inventario
    def __init__(self):
        self.productos = {}

    #metodo para agregar un producto
    def agregar_producto(self, producto):
        if producto.id not in self.productos:
            self.productos[producto.id] = producto
        else:
            print("Ya existe un producto con este ID en el inventario.\n")
        
    #metodo para eliminar producto
    def eliminar_producto(self, id):
        if id in self.productos:
            del self.productos[id]
        else:
            print("El producto con este ID no existe en el inventario.\n")
    
    #metodo para actualizar producto por id
    def actualizar_producto(self, id, nuevo_producto):
        if id in self.productos:
            self.productos[id] = nuevo_producto
        else:
            print("El producto con este ID no existe en el inventario.\n")

    #metodo para buscar producto por id
    def buscar_producto(self, id):
        if id in self.productos:
            print ("El producto es:\n", self.productos[id]) 
            print("\n")
        else:
            print("El producto con este ID no existe en el inventario.\n")

    #metodo para listar todos los productos del inventario
    def listar_producto(self):
        for producto in self.productos.values():
            print(producto)

    #metodo calcular la cantidad total de productos
    def cantidad_total_productos(self):
        return sum(producto.cantidad for producto in self.productos.values())

    #metodo para calcular el valor total del inventario
    def valor_total_inventario(self):
        return sum(producto.cantidad * producto.precio for producto in self.productos.values())
```

```python
#Instancia inicial de productos para ser agregados en el inventario
producto1 = Producto("1", "Producto 1", "Descripción del Producto 1", 10, 100)
producto2 = Producto("2", "Producto 2", "Descripción del Producto 2", 20, 200)
producto3 = Producto("3", "Producto 3", "Descripción del Producto 3", 30, 300)

#se crea un objeto inventario y se agregan los productos creados anteriormente
inventario = Inventario()
inventario.agregar_producto(producto1)
inventario.agregar_producto(producto2)
inventario.agregar_producto(producto3)

#menu principal para el usuario
while True:
    print("Bienvenido a nuestro inventario:")
    print("Si quiere agregar un nuevo producto ingrese 1")
    print("Si quiere eliminar un producto ingrese 2")
    print("Si quiere actualizar un producto ingrese 3")
    print("Si quiere buscar un producto ingrese 4")
    print("Si quiere ver todos los productos ingrese 5")
    print("Si quiere salir del programa ingrese 0")
    opcion = input("Ingrese la opción seleccionada:")
    print("\n")
    
    #manejo de excepcion para controlar que se ingresa una opcion correcta
    try:
        if(opcion == '1'):
            # Opcion para agregar producto al inventario. Solicitar los datos del producto 
            id = input("Ingrese el id del producto:")
            nombre = input("Ingrese el nombre del producto:")
            descripcion = input("Ingrese la descripcion del producto:")
            cantidad = int(input("Ingrese la cantidad del producto:"))
            precio = int(input("Ingrese el precio del producto:"))
    
            #agregar el producto al inventario
            inventario.agregar_producto(Producto(id, nombre, descripcion, cantidad, precio))
    
        elif (opcion == '2'):
            # Opcion para eliminar producto del inventario. Solicitar id del producto 
            id = input("Ingrese el id del producto a eliminar:")
            inventario.eliminar_producto(id)
            
        elif (opcion == '3'):
            # Opcion para actualizar un producto del inventario. Solicitar los datos del producto 
            id = input("Ingrese el id del producto a actualizar:")
            nombre = input("Ingrese el nombre del producto:")
            descripcion = input("Ingrese la descripcion del producto:")
            cantidad = int(input("Ingrese la cantidad del producto:"))
            precio = int(input("Ingrese el precio del producto:"))
            inventario.actualizar_producto(id, Producto(id, nombre, descripcion, cantidad, precio))
            
        elif(opcion == '4'):
            # Opcion para buscar un producto en el inventario. Solicitar id del producto 
            id = input("Ingrese el id del producto a buscar:")
            inventario.buscar_producto(id)
            
        elif(opcion == '5'):
            # Opcion para mostrar todos los productos del inventario, la cantidad total de productos y el valor del inventario 
            print("Los productos en el inventario son los siguientes:")
            inventario.listar_producto()
            print("La cantidad total de productos es:", inventario.cantidad_total_productos())
            print("El valor total del inventario es:", inventario.valor_total_inventario())
        
        elif(opcion == '0'):
            #opcion para salir del bucle
            print("Ha elegido salir del programa. Adios!")
            break
        else:
            print("La opción ingresada no es válida\n") 
        
    except ValueError:
        print("La opción ingresada no es válida\n")
    
```
